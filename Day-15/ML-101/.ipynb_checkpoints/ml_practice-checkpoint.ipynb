## Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

## Load Dataset
-It is better to keep the dataset in the same folder in which you have your notebook, otherwise you have to enter the complete path

```python
# Load dataset
df = pd.read_csv("Aadi_data.csv")
df.head()
```

```python
# Take relevant data
workshop_data = df[["age", "weight", "height"]]
workshop_data.head()
```

```python
x = workshop_data.iloc[:, :1].values #get a copy of dataset exclude last column
y = workshop_data.iloc[:, 1].values # get array of dataset in column 1st
```

```python
x
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=1/3, random_state=0)
```

```python
# Fitting Similar Linear Regression to the Training set
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# Visualizing the Traning set results
viz_train = plt
viz_train.scatter(X_train, y_train, color='red')
viz_train.plot('Age vs weight')


# Visualizing the Test set results
```
